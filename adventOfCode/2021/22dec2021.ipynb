```python
#task 1
def fixRange(inRange):
    outRange = []
    for count in range(3):
        if (inRange[count][0]<-50 and inRange[count][1]<-50) or (inRange[count][0]>50 and inRange[count][1]>50):
            return [(0,0),(0,0),(0,0)]
        if inRange[count][0]<-50:
            r1 = -50
        else:
            r1 = inRange[count][0]
        if inRange[count][1]>50:
            r2 = 50
        else:
            r2 = inRange[count][1]
        outRange.append((r1,r2))
    return outRange

def updateGrid(mode, grid, rng):
    for a in range(rng[0][0],rng[0][1]+1):
        for b in range(rng[1][0],rng[1][1]+1):
            for c in range(rng[2][0],rng[2][1]+1):
                grid[a][b][c] = 1 if mode=="on" else 0
    return grid

def execIns(grid, ins):
    typ = ins.split(' ')[0]
    ranges = [ele.split('=')[1] for ele in ins.split(' ')[1].split(',')]
    ranges = [tuple([int(a) for a in ele.split('..')]) for ele in ranges]   
    ranges = fixRange(ranges)
    if (0,0) in ranges:
        return grid
    return updateGrid(typ, grid, ranges)

def countOn(grid):
    count = 0
    for a in grid:
        for b in a:
            for c in b:
                if c==1:
                    count += 1
    return count

with open("22dec2021input.txt", "r") as f:
    inss = f.read().split('\n')

grid = [[[0 for a in range(101)] for b in range(101)] for c in range(101)]

for ins in inss:
    grid = execIns(grid, ins)

print(countOn(grid))    
```

```python
class Ranges:
    
    def __init__(self, rngs=[]):
        if type(rngs)==type(self):
            self.rngs = rngs.rngs[:]
            return
        if type(rngs)==list and len(rngs)!=0 and type(rngs[0]) == type(self):
            self.rngs = [ele.rngs[0] for ele in rngs]
            return
        if not (type(rngs) == list and len([ele for ele in rngs if type(ele)==tuple and len(ele)==2 and ele[0]<ele[1]])==len(rngs)):
            print("Invalid parameter passed to Range()!")
            self.rngs = []
            return
        self.rngs = [ele for ele in rngs]
    
    def __bool__(self):
        if self.rngs == []:
            return False
        else:
            return True
    
    def simplify(self):
        self.rngs = sorted(self.rngs, key=lambda x:x[0])
        i=0
        while i < len(self.rngs)-1:
            if not (self.rngs[i][0] > self.rngs[i+1][1] or self.rngs[i+1][0] > self.rngs[i][1]):
                minP = self.rngs[i][0] if self.rngs[i][0]<self.rngs[i+1][0] else self.rngs[i+1][0]
                maxP = self.rngs[i][1] if self.rngs[i][1]>self.rngs[i+1][1] else self.rngs[i+1][1]
                self.rngs.pop(i)
                self.rngs.pop(i)
                self.rngs.insert(i,(minP,maxP))
                i = -1
            i += 1
        
    def __add__(self, other):
        outRange = Ranges()
        outRange.rngs = self.rngs+other.rngs
        outRange.simplify()
        return outRange
    
    def __EQ__(self, other):
        if self.rngs == other.rngs:
            return True
        else:
            False
    
    def __xor__(self, other):
        if not other or not self:
            return Ranges()
        intersections = []
        for a in self.rngs:
            for b in other.rngs:
                if not (a[0] >= b[1] or b[0] >= a[1]):
                    minP = a[0] if a[0]>b[0] else b[0]
                    maxP = a[1] if a[1]<b[1] else b[1]
                    intersections.append((minP,maxP))
        intersections = sorted(intersections, key=lambda x:x[0])
        return Ranges(intersections)
    
    def __str__(self):
        if len(self.rngs)==1:
            return str(self.rngs).strip('[').strip(']')
        else:
            return str(self.rngs)
    
    def __repr__(self):
        return repr(self.rngs)
        
    def __sub__(self, other):
        if not other:
            return Ranges(self)
        (onRange, remRange) = (self.rngs,other.rngs)
        minP = onRange[0][0] if onRange[0][0] < remRange[0][0] else remRange[0][0]
        maxP = onRange[-1][1] if onRange[-1][1] > remRange[-1][1] else remRange[-1][1]
        oppSubRange = Ranges()
        for i in range(len(other.rngs)-1):
            oppSubRange.rngs.append((other.rngs[i][1],other.rngs[i+1][0]))
        if minP!=other.rngs[0][0]:
            oppSubRange.rngs.insert(0,(minP, other.rngs[0][0]))
        if maxP!=other.rngs[-1][1]:
            oppSubRange.rngs.append((other.rngs[-1][1], maxP))
        oppSubRange.rngs = [ele for ele in oppSubRange.rngs if ele[0]<ele[1]]
        outRange = self^oppSubRange
        return outRange        
    
    def countEle(self) -> int:
        self.simplify()
        count = 0
        for ele in self.rngs:
            count += ele[1]-ele[0]
        return count
    
class MultiDimRanges:
    
    def __init__(self, param=None):
        if param==None:
            self.level=1
            self.rngs = []
            self.subrngs = []
        elif type(param) == type(self):
            self.level = param.level
            self.rngs = [ele for ele in param.rngs]
            self.subrngs = [ele for ele in param.subrngs]
        else:
            self.rngs = [Ranges([ele]) for ele in param[0]]
            self.subrngs = [ele for ele in param[1]]
            self.level = param[1][0].level+1 if param[1]!=[] else 1
    
    def __EQ__(self, other):
        if self.level==other.level and self.rngs == other.rngs and self.subrngs==other.subrngs:
            return True
        else:
            return False
    
    def __bool__(self):
        return False if self.rngs == [] else True 

    def simplify(self):
        if self.level==1:
            temp = Ranges(self.rngs)
            temp.simplify()
            self.rngs = [Ranges([ele]) for ele in temp.rngs]
            return
        self.rngs,self.subrngs = (list(t) for t in zip(*sorted(zip(self.rngs, self.subrngs), key=lambda x: x[0].rngs[0][0])))
        i=0
        while i < len(self.rngs)-1:
            union = self.rngs[i]+self.rngs[i+1]
            intersection = self.rngs[i]^self.rngs[i+1]
            if repr(self.subrngs[i])==repr(self.subrngs[i+1]) and len(union.rngs)==1:
                (self.rngs.pop(i),self.subrngs.pop(i),self.rngs.pop(i))
                (self.rngs.insert(i,union), self.subrngs.insert(i,self.subrngs.pop(i)))
                i = -1
            elif intersection:
                (newUpper,newLower) = ([],[])
                lowerUnion = self.subrngs[i]+self.subrngs[i+1]
                (newUpper.append(intersection),newLower.append(lowerUnion))
                if union-self.rngs[i]:
                    (newUpper.append(union-self.rngs[i]),newLower.append(self.subrngs[i+1]))
                if union-self.rngs[i+1]:
                    (newUpper.append(union-self.rngs[i+1]),newLower.append(self.subrngs[i]))
                newUpper,newLower = (list(t) for t in zip(*sorted(zip(newUpper, newLower), key=lambda x: x[0].rngs[0][0])))
                [(self.rngs.pop(i),self.subrngs.pop(i)) for count in range(2)]
                [self.rngs.insert(i,ele) for ele in newUpper]
                [self.subrngs.insert(i,ele) for ele in newLower]
                self.rngs,self.subrngs = (list(t) for t in zip(*sorted(zip(self.rngs, self.subrngs), key=lambda x: x[0].rngs[0][0])))
                i = -1
            i += 1
    
    def __add__(self, other):
        out = MultiDimRanges()
        if not self:
            return other
        elif not other:
            return self
        elif self.level!=other.level:
            return out
        else:
            out.level = self.level
            out.rngs = self.rngs + other.rngs
            out.subrngs = self.subrngs + other.subrngs
            out.simplify()
            return out
    
    def __xor__(self, other):
        intersections = []
        if self.level!=other.level:
            return MultiDimRanges()
        for count1 in range(len(self.rngs)):
            for count2 in range(len(other.rngs)):
                intersections.append((self.rngs[count1]^other.rngs[count2], self.subrngs[count1]^other.subrngs[count2] if self.level!=1 else 1))
        intersections = [ele for ele in intersections if ele[0] and ele[1]]
        if self.level==1:
            return MultiDimRanges([[ele[0] for ele in intersections],[]])
        else:
            return MultiDimRanges([[ele[0] for ele in intersections],[ele[1] for ele in intersections]])
  
    def __sub__(self, other):
        if self.level!=other.level:
            return MultiDimRanges()
        elif self.level==1:
            return MultiDimRanges([[ele for ele in (Ranges(self.rngs)-Ranges(other.rngs)).rngs],[]])
        subs = []
        for count1 in range(len(self.rngs)):
            safeRange = [Ranges([ele]) for ele in (self.rngs[count1] - (self.rngs[count1]^Ranges(other.rngs))).rngs]
            subs.extend([(ele,MultiDimRanges(self.subrngs[count1])) for ele in safeRange])
            for count2 in range(len(other.rngs)):
                subs.append((self.rngs[count1]^other.rngs[count2], self.subrngs[count1]-other.subrngs[count2]))
        subs = [ele for ele in subs if ele[0] and ele[1]]
        out = MultiDimRanges([[ele[0] for ele in subs],[ele[1] for ele in subs]])
        out.simplify()
        return out
    
    def __get_printable(self, spaceCount = 0):
        out = ""
        for i in range(len(self.rngs)):
            out += " "*spaceCount + self.rngs[i].__str__()
            if self.level > 1:
                out += "\n" + " "*spaceCount + self.subrngs[i].__get_printable(spaceCount + 4)
            if i+1!=len(self.rngs):
                out += "\n"
        return out
    
    def __str__(self):
        return self.__get_printable()
        
    def countEle(self) -> int:
        self.simplify()
        count = 0
        for i in range(len(self.rngs)):
            if self.level>1:
                count += self.rngs[i].countEle() * self.subrngs[i].countEle()
            else:
                count += self.rngs[i].countEle()
        return count
    
    def __repr__(self):
        return "%s%s"%(repr(self.rngs), "" if self.level==1 else ":"+repr(self.subrngs))
        
```
